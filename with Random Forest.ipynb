```python
import pandas as pd
```

```python
# importing libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
# data sourcing (data sourced from different schools of 10th passed out students)
df = pd.read_csv(r"C:\Users\badya\Downloads\10th result Project.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
def recommend_stream(row):
    maths = row['Maths']
    science = row['Science']
    english = row['English']
    
    avg = (maths + science + english) / 3

    if science >= 80 and maths >= 85:
        return 'Science (Non-Medical)'
    
    elif science >= 80 and english >= 70:
        return 'Science (Medical)'
    
    elif maths >= 60:
        return 'Commerce'
    
    else:
        return 'Arts'


df['Recommended_Stream'] = df.apply(recommend_stream, axis=1)
```

```python
df
```

```python
df.shape
```

```python
df['Recommended_Stream'] = df['Recommended_Stream'].map({'Science (Medical)': 0, 'Science (Non-Medical)': 1, 'Commerce':2, 'Arts':3})
```

```python
x = df[['Science','English','Maths']]
```

```python
y = df['Recommended_Stream']
```

```python
y.isnull().sum()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
model = RandomForestRegressor(n_estimators=100, max_depth=4, random_state=42)
```

```python
model.fit(x_train_scaled,y_train)
```

```python
# prediction on test data
y_pred = model.predict(x_test_scaled)
```

```python
r2_score(y_test,y_pred)
```

```python
# prediction on training data
x_pred = model.predict(x_train_scaled)
```

```python
r2_score(y_train,x_pred)
```
